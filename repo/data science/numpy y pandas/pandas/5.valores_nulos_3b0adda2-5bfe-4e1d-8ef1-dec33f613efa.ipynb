### Importar librerías

```python
import pandas as pd 
import numpy as np 
```

### Crear diccionario y DataFrame

```python
dict = {'Col1':[1,2,3,np.nan],
 'Col2':[4,np.nan,6,7],
 'Col3':['a','b','c',None]}
```

```python
df = pd.DataFrame(dict)
```

```python
df
```

### Ver valores nulos

```python
df.isnull()
```

```python
df.isnull()*1
```

### Llenar valores nulos

```python
df.fillna('Missing')
```

```python
df.fillna(df.mean()) # df.mean() es llenar con un valor como la media 
#  el warning dice que no puede sacar media a valores nan, pero igual la hace
```

### Interpolar

```python
df.interpolate() # asume el valor de una serie en los valores nulos
```

### Borrar valores nulos

```python
df.dropna()
```
