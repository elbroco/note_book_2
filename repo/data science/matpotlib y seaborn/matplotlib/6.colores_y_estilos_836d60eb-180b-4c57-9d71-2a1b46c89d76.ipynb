```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

### Style

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color = 'green', alpha=0.8, linewidth=5, linestyle='-',marker='o'
,markersize=10, markerfacecolor='#CCFFCC')
ax.plot(x,x+2,color = 'blue', linewidth=8 , linestyle='--',marker='8')
ax.plot(x,x+3,color = '#66FF66', linewidth=3 ,linestyle='dashed',marker='v')
ax.plot(x,x+4,color = '#CCFFCC', linewidth=2 ,linestyle=':',marker='P')
```

```python
import matplotlib.pyplot as plt

# Define the function y = (3 + x)^3
def f(x):
    return (3 + x)**3

# Create a range of x values
x = np.linspace(-10, 10, 400)
y = f(x)

# Loop through all available styles in Matplotlib
for style in plt.style.available:
    plt.style.use(style)
    fig, ax = plt.subplots(figsize=(2, 2))
    ax.plot(x, y)
    ax.set_title(style, size=20)
    plt.show()  # Display the plot
  
  
```
