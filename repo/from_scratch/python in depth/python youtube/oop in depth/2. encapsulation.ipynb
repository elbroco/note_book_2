```python
#  my solution

class Guest:
    guests = []  # Class variable to store all guests
    
    def __init__(self, first_n, last_n, rank, age):
        self.first_n = first_n.capitalize()
        self.last_n = last_n.capitalize()
        self.rank = int(rank)
        self.age = int(age)
        
    def adding_g(self):
        guest = [self.first_n, self.last_n, self.rank, self.age]
        Guest.guests.append(guest)
        print(f'{self.first_n} has been added.')
        
    @classmethod
    def num_of_guests(cls):
        return len(cls.guests)
    
    @classmethod
    def display_guests(cls):
        for guest in cls.guests:
            print('-----------')
            print(f'First name: {guest[0]}')
            print(f'Last name: {guest[1]}')
            print(f'Age: {guest[3]}')
            print('-----------')
            
    @classmethod
    def loyalty(cls):
        for i in cls.guests:
            if i[2] >= 10:
                print(f'guest {i[0]} is a gold member')
            
        
            
    
        

def rep(n):
    for _ in range(n):
        getting_info()

def getting_info():
    f_name = input('Enter a first name: ')
    l_name = input('Enter a last name: ')
    rank = input('Enter a rank: ')
    age = input('Enter an age: ')

    run = Guest(f_name, l_name, rank, age)
    run.adding_g()

n = int(input('How many guests would you like to add?: '))
rep(n)

# After adding guests
total_guests = Guest.num_of_guests()
print(f'Total number of guests: {total_guests}')

# Display all guests' information
Guest.display_guests()

Guest.loyalty()
```

```python
class Guest:
    def __init__(self, first, last, rank, age):
        self.last_name = last
        self.first_name = first
        self.rank = int(rank)
        self.age = int(age)
        
    def guest_info(self, all_guests):
        for guest in all_guests:
            print(guest.first_name, guest.last_name, 'age:', guest.age)
            
    def loyalty_program(self, all_guests):
        gold_members = [guest for guest in all_guests if guest.rank >= 10]
        if gold_members:
            print('Gold members:')
            for member in gold_members:
                print(f'\tGuest: {member.first_name} {member.last_name}')
    
    def guest_avg(self, all_guests):
        total_age = sum(guest.age for guest in all_guests)
        avg_age = total_age / len(all_guests)
        print('Average customer age:', avg_age)
        
all_guests = []
num_guests = int(input('Enter number of guests: '))
for i in range(num_guests):
    data = input('First name/Last name/Rank/Age: ').split('/')
    guest = Guest(data[0], data[1], data[2], data[3])
    all_guests.append(guest)
    
guest = all_guests[0]  # Example: Assuming we want to work with the first guest in the list
guest.loyalty_program(all_guests)
guest.guest_info(all_guests)
guest.guest_avg(all_guests)
```
